
# Linear Regression from Scratch using Python

This notebook implements linear regression without using
machine learning libraries, demonstrating the mathematical
foundation of supervised learning.


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Dataset

We generate a simple linear dataset:
y = mx + c + noise

This mimics experimental or physical data.


```python
np.random.seed(42)

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Generated Data")
plt.grid(True)
plt.show()
```

#As x increases, y increases. The output variable y increases approximately linearly with the input variable x. The data is not perfectly straight. 

The points are not on a perfect line. The points are not on a perfect line, Because of this term: np.random.randn(100, 1) This adds random noise.

Interpretation

The data contains measurement noise, similar to real experimental or physical data.

## Linear Regression Model

The model is:
y = wx + b

We minimize Mean Squared Error (MSE):
MSE = (1/n) Σ (y_pred − y)²


```python
w = np.random.randn()
b = np.random.randn()

learning_rate = 0.1
n_epochs = 1000
```

```python
n = len(X)

for epoch in range(n_epochs):
    y_pred = w * X + b
    
    dw = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)
    
    w = w - learning_rate * dw
    b = b - learning_rate * db

w, b
```

```python
y_pred = w * X + b

plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression from Scratch")
plt.grid(True)
plt.show()
```



> *The red regression line closely follows the central trend of the scattered data points, showing that the learned linear model accurately captures the underlying linear relationship despite the presence of noise.*
